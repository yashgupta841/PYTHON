```python
pi=22/7
omega = 10
Y = np.sin(2 * pi * omega * X)
Y 
#Y.shape
#Y.ndim
Y[0:10]
```

```python
plt.plot(X,Y, color="green",linewidth=1,marker="*")
plt.title(f"Sine wave $\omega ={omega}$")
plt.xlabel("Time in (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.savefig("test1.png")
plt.plot(X,Y)
```

```python
PI = 3.14
start = 1
end = 3600
stepsize = 10
X = np.arange(start, end, stepsize)

omega = 1
omega2 = 3

Y=np.sin(2 * PI * omega * X)
Y2 = 2 * np.sin(2 * PI * omega2 * X)

plt.figure()
plt.plot(X,Y, label = "label1", marker="*")
plt.plot(X,Y2, label = "label 2", marker="*")
plt.legend()
```

```python
#$$ y1=2/sin/2/pi/omega/X
PI = 22/7
start = 1
end = 1*60*60
noOfpoints = 100
X = np.linspace(start, end, noOfpoints)

omega = 1
omega2 = 3

Y1=np.sin(2 * PI * omega * X)
Y2 = 2 * np.sin(2 * PI * omega2 * X)

plt.figure()

plt.title("Figure 1")
plt.xlabel("time (s)")
plt.ylabel("Amplitude")
plt.plot(X,Y1, label = "Y1", color="red")
plt.plot(X,Y2, label = "Y2", linewidth=4)
plt.legend()
plt.grid()
filename = "show.png"
plt.savefig(filename)
```
